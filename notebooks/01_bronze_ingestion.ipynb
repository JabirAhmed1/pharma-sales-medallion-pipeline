# Business Context

Acme Pharmaceuticals has delivered four raw datasets representing
prescriptions, providers, products, and sales regions.

The objective of this notebook is to ingest these files into the
Bronze layer while preserving the original source data.

No business logic or cleaning is performed in this stage.


```python
from pyspark.sql import functions as F
```

```python
prescriptions_df = spark.table("raw_prescriptions")
products_df = spark.table("raw_products")
providers_df = spark.table("raw_providers")
regions_df = spark.table("raw_regions")
```

```python
print("Presecription Rows:", prescriptions_df.count())
print("Products Rows:", products_df.count())
print("Providers Rows:", providers_df.count())
print("Regions Rows:", regions_df.count())
```

```python
products_df.printSchema()
providers_df.printSchema()
regions_df.printSchema()
```

```python
display(products_df)
display(providers_df)
display(regions_df)
```

```python
prescriptions_df.write.format("delta").mode("overwrite").saveAsTable("bronze_perscriptions")
products_df.write.format("delta").mode("overwrite").saveAsTable("bronze_products")
providers_df.write.format("delta").mode("overwrite").saveAsTable("bronze_providers")
regions_df.write.format("delta").mode("overwrite").saveAsTable("bronze_regions")
```
